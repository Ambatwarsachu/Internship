2) Write a python program to scrape details of all the posts from https://www.patreon.com/coreyms .Scrape the
heading, date, content and the likes for the video from the link for the youtube video from the post.

```python
head = 'https://www.patreon.com/coreyms'


def youtube_posts(head):
    res = requests.get(head)
    soup = BeautifulSoup(res.content, 'html.parser')
    heading = [i.text for i in soup.find_all('div', class_="sc-bBHxTw jIEOUn")]
    date = [i.text for i in soup.find_all('div', style_="display: flex;")]
    content = [i.text for i in soup.find_all('div', class_="sc-pn81yk-1 fIwkbU")]
    linkes = [i.text for i in soup.find_all('div', clas_="sc-jrQzAO iubCOz")]
    df = pd.DataFrame({'Heading':heading,
                      'Date':date,
                      'Content':content,
                       'likes':linkes
                      })
    return df

youtube_data = youtube_posts(head)

youtube_data
```

4) Write a python program to scrape first 10 product details which include product name , price , Image URL from
https://www.bewakoof.com/bestseller?sort=popular .



```python
link = requests.get('https://www.bewakoof.com/bestseller?sort=popular%20')
soups = BeautifulSoup(link.content, 'html.parser')


img_url = []
for j in soups.find_all('img', class_="productImgTag")[:10]:
    img_url.append(j.get("src"))
img_url


product_name = []
for i in soups.find_all('div', class_="productNaming bkf-ellipsis")[:10]:
    product_name.append(i.text)
product_name

price = []
for k in soups.find_all('div', class_="loyaltyPriceBox")[:10]:
    price.append(k.text)
price

df = pd.DataFrame({'Product_Name':product_name, 'Price':price, 'Link':img_url})
df
```

5) Please visit https://www.cnbc.com/world/?region=world and scrap-
 a) headings
b) date
c) News link

```python
 
from bs4 import BeautifulSoup
import requests
import pandas as pd

url = 'https://www.cnbc.com/world/?region=world'
def scrape_news_data(url):
    response = requests.get(url)
    
    soup = BeautifulSoup(response.content, 'html.parser')

    headings = [i.text for i in soup.find_all('a', class_="LatestNews-headline")]

    times = [j.text for j in soup.find_all('time', class_="LatestNews-timestamp")]

    links = [k.get("href") for k in soup.find_all('a', class_="LatestNews-headline")]

    df = pd.DataFrame({'Headings': headings, 'Time': times, 'Link': links})

    return df

news_data = scrape_news_data(url)

print(news_data)
```

6. Please visit https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloadedarticles/ and scrap
a) Paper title
b) date
c) Author

```python
paper = requests.get('https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded-articles/')
paper

soup = BeautifulSoup(paper.content, 'html.parser')

paper_title = []
for i in soup.find_all('h2', class_="h5 article-title"):
    paper_title.append(i.text)

date = []
for i in soup.find_all('p', class_="article-date"):
    date.append(i.text)

author = []
for j in soup.find_all('p', class_="article-authors"):
    author.append(j.text)

df = pd.DataFrame({
    'Paper Title': paper_title,
    'Date':date,
    'Author':author
})

df
```
